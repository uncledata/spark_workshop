## How to publish my changes to main then?

```python
%%sql
SELECT * FROM staging.aggregated_taxi_per_vendor_date.refs 
```

```python
%%sql
select * from staging.aggregated_taxi_per_vendor_date.snapshots
```

```python
query = f"""
SELECT snapshot_id
FROM staging.aggregated_taxi_per_vendor_date.refs
WHERE name = 'manual_hot_fix'
"""
wap_snapshot_id = spark.sql(query).head().snapshot_id
```

```python
wap_snapshot_id
```

```python
publish_query = f"CALL system.cherrypick_snapshot('staging.aggregated_taxi_per_vendor_date', {wap_snapshot_id})"

%sql $publish_query
```
